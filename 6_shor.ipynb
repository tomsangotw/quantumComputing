```python
import qiskit
import numpy as np
import matplotlib.pyplot as plt
```

```python
""" 
    Qiskit
    https://qiskit.org/documentation/getting_started.html
"""
from qiskit import(
  QuantumCircuit,
  execute,
  Aer)
from qiskit.visualization import plot_histogram
```

```python
"""
    Factoring as Period Finding Problem
"""
# f(x) = a^x mod N , a 與 N 互質 
# 目標變成找到一個 non-zero r 讓 a^r mod N = 1 且 
#  (1) r 是偶數 
#  (2) a(r/2) + 1 != 0 mod N 
# 因為可以拆成 (a(r/2) + 1) (a(r/2) - 1) = 0 mod M 
# 則 p 與 q 會是倍數項 -> 靠 GCD 抓出來 
#  p = GCD(a(r/2) + 1, M) 
#  q = GCD(a(r/2) - 1, M) 

# (1) r = 12 偶數 
# (2) r^6 + 1 = 30 (mod 35) 不為 0 則 
# p = GCD(30, 35) = 5 
# q = GCD(28, 35) = 7 

N = 35
a = 3

# Calculate the plotting data
xvals = np.arange(35)
yvals = [np.mod(a**x, N) for x in xvals]
x = [i for i in range(35)]
print(x)
print(yvals)

# Use matplotlib to display it nicely
fig, ax = plt.subplots()
ax.plot(xvals, yvals, linewidth=1, linestyle='dotted', marker='x')
ax.set(xlabel='$x$', ylabel='$%i^x$ mod $%i$' % (a, N),
       title="Example of Periodic Function in Shor's Algorithm")
try: # plot r on the graph
    r = yvals[1:].index(1) +1 
    plt.annotate(s='', xy=(0,1), xytext=(r,1), arrowprops=dict(arrowstyle='<->'))
    plt.annotate(s='$r=%i$' % r, xy=(r/3,1.5))
except:
    print('Could not find period, check a < N and have no common factors.')
    
```

```python
"""
    Shor's algorithm:
    https://qiskit.org/textbook/ch-algorithms/shor.html
"""
# Input: N = p * q, p 與 q 都是質數 
# Output: p, q 
# 實際上變成找周期的問題，因為可藉由週期線性解出質因數分解 

def c_amod15(a, power):
    """Controlled multiplication by a mod 15"""
    if a not in [2,7,8,11,13]:
        raise ValueError("'a' must be 2,7,8,11 or 13")
    U = QuantumCircuit(4)        
    for iteration in range(power):
        if a in [2,13]:
            U.swap(0,1)
            U.swap(1,2)
            U.swap(2,3)
        if a in [7,8]:
            U.swap(2,3)
            U.swap(1,2)
            U.swap(0,1)
        if a == 11:
            U.swap(1,3)
            U.swap(0,2)
        if a in [7,11,13]:
            for q in range(4):
                U.x(q)
    U = U.to_gate()
    U.name = "%i^%i mod 15" % (a, power)
    c_U = U.control()
    return c_U

def qft_dagger(n):
    """n-qubit QFTdagger the first n qubits in circ"""
    qc = QuantumCircuit(n)
    # Don't forget the Swaps!
    for qubit in range(n//2):
        qc.swap(qubit, n-qubit-1)
    for j in range(n):
        for m in range(j):
            qc.cu1(-np.pi/float(2**(j-m)), m, j)
        qc.h(j)
    qc.name = "QFT†"
    return qc

simulator = Aer.get_backend('qasm_simulator')

# 以 a = 7, N = 15 為例
# x: 1, 2,  3, 4
# f: 7, 4, 13, 1
a = 7
circuit = QuantumCircuit(3 + 4, 3) # 需要 3 qubit + 4 control qubit
circuit.h([0, 1, 2]) 
circuit.x(6) #(=Not gate) 讓 qubit 6 (第 7 個) 的輸入是 1
# 因為所有獨特的 eigenstate 加總起來，會是 |1> 
# 所以我們用 Quantum Phase Estimation 測量 |1> 的 phase 會得到 M = s/r  
circuit.barrier()

# QPE 的部分, cotrolled-U
for q in range(3):
    circuit.append(c_amod15(a, 2**q), [q] + [i + 3 for i in range(4)])
    circuit.barrier()

# Inverse-QFT
circuit.append(qft_dagger(3), range(3)) # 得到 phase
circuit.barrier()

# Map the quantum measurement to the classical bits
circuit.measure([0, 1, 2], [0, 1, 2])

# Execute the circuit on the qasm simulator
job = execute(circuit, simulator)

# Grab results from the job
result = job.result()

# Returns counts
counts = result.get_counts(circuit)
print("Result are:",counts)

# Draw the circuit
circuit.draw(output='mpl')
# 順序 q3, q2, q1 ...

# bin 
# 0b 010 = 2 => 2/8 = 1/4 (r = 4) 
# 0b 100 = 1 => 4/8 = 1/2 (r = 2) 
# 0b 000 = 0 => 0/8 = 0/1 (r = 1) 
# 0b 110 = 6 => 6/8 = 3/4 (r = 4) 
# 得到 r = 4
# 7^2 + 1 = 5 (mod 15)
# 7^2 - 1 = 3 (mod 15)
```
